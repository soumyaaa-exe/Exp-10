```python
#Create DataFrame
df=pd.DataFrame({'Name':['Alex','Tom','Steve'],'Marks':[34,29,42]})
print(df)
```

```python
print(df['Name'])
print('')
print(df['Marks'])
```

```python
print(df.loc[0,'Name'])
print('')
print(df.iloc[2,1])
```

```python
#Add a new column
df['Grade']=['A','B','C']
print(df)
```

```python
#Delete a Column using python library
del df['Grade']
print(df)
```

```python
#Update data in DataFrame
df.loc[0,'Marks']=45
print(df)
print(' ')
df.iloc[1,1]=36
print(df)
```

```python
#Delete a Column using Pandas library (inplace overwrites data) (Axis defines row (0) or column (1))
df.drop('Grade',axis=1,inplace=True)
print(df)
```

```python
df1=df.drop('Grade',axis=1)
print(df)
print(' ')
print(df1)
```

```python
#Perform basic Statistical Analysis
mean1=df['Marks'].mean()
print(mean1)

median1=df['Marks'].median()
print(median1)
max1=df['Marks'].max()
print(max1)
min1=df['Marks'].min()
print(min1)
```

```python
#Apply Condition (Filtering)
df[df['Marks']>40]
```
